## Exploratory Data Analysis (EDA)

This exploratory data analysis examines a global CO₂ and greenhouse gas emissions dataset covering multiple countries and a long historical time span. The analysis focuses on understanding the structure, scale, and availability of emissions data before any modelling or predictive techniques are applied.


```python
# Group the data by year and calculate total global CO2 emissions
global_co2 = df.groupby("year")["co2"].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(global_co2["year"], global_co2["co2"])
plt.xlabel("Year")
plt.ylabel("Global CO₂ Emissions")
plt.title("Global CO₂ Emissions Over Time")
plt.show()
```

### Global CO₂ Emissions Over Time

This figure shows the total global CO₂ emissions aggregated across all countries from 1750 to the present. The data indicates very low emissions prior to the industrial revolution, followed by a gradual increase during the 19th century. A sharp acceleration is observed after the mid-20th century, reflecting industrialisation, population growth, and increased fossil fuel consumption. This trend aligns with established findings in climate change literature, suggesting that the dataset reliably captures long-term global emission patterns.

```python
df_modern = df[df["year"] >= 1960]
df_modern["co2"].isna().sum(), df_modern.shape
```

```python
top_emitters = (
    df_modern.groupby("country")["co2"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_emitters
```

```python
df_countries = df_modern[df_modern["iso_code"].str.len() == 3]
```

```python
top_emitters_countries = (
    df_countries.groupby("country")["co2"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_emitters_countries
```
