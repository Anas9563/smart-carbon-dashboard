```python
# --- Data Cleaning ---
# Based on the inspection above, the following cleaning steps are applied
# to create a clean base dataset for use across all pages and notebooks

df_clean = df.copy()

# 1. Filter to post-1960
df_clean = df_clean[df_clean["year"] >= 1960]

# 2. Check and remove duplicates
print(f"Duplicates found: {df_clean.duplicated().sum()}")
df_clean = df_clean.drop_duplicates()

# 3. Drop rows where CO2 is null
rows_before = len(df_clean)
df_clean = df_clean.dropna(subset=["co2"])
print(f"Rows removed due to missing CO2: {rows_before - len(df_clean)}")

# 4. Sort by country and year
df_clean = df_clean.sort_values(["country", "year"])

# 5. Reset index
df_clean = df_clean.reset_index(drop=True)

# 6. Confirm cleaning results
print(f"\nCleaning Summary:")
print(f"Original rows: {len(df)}")
print(f"Cleaned rows: {len(df_clean)}")
print(f"Countries: {df_clean['country'].nunique()}")
print(f"Year range: {df_clean['year'].min()} - {df_clean['year'].max()}")
print(f"Missing CO2 values remaining: {df_clean['co2'].isna().sum()}")

# 7. Save cleaned dataset
df_clean.to_csv("../data/cleaned_co2.csv", index=False)
print("\nCleaned dataset saved to data/cleaned_co2.csv")
```

## Exploratory Data Analysis (EDA)

This exploratory data analysis examines a global CO2and greenhouse gas emissions dataset covering multiple countries and a long historical time span. The analysis focuses on understanding the structure, scale, and availability of emissions data before any modelling or predictive techniques are applied.

```python
# Group the data by year and calculate total global CO2 emissions
global_co2 = df_clean.groupby("year")["co2"].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(global_co2["year"], global_co2["co2"])
plt.xlabel("Year")
plt.ylabel("Global CO₂ Emissions")
plt.title("Global CO₂ Emissions Over Time")
plt.show()
```

### Global CO2Emissions Over Time

This figure shows the total global CO2 emissions aggregated across all countries from 1960 to the present. The data indicates very low emissions prior to the industrial revolution, followed by a gradual increase during the 19th century. A sharp acceleration is observed after the mid-20th century, reflecting industrialisation, population growth, and increased fossil fuel consumption. This trend aligns with established findings in climate change literature, suggesting that the dataset reliably captures long-term global emission patterns.
